```python
import pandas as pd

CLEAN_PATH = '../../utaut2_dataset/new_result.xlsx'
SUBSAMPLE_PATH = '../../utaut2_dataset/utaut2_subsample.xlsx'

df = pd.read_excel(CLEAN_PATH)
print('Полная очищенная выборка:', df.shape)
```

```python
cols = df.columns.tolist()

# --- UTAUT2: формирование аналитической подвыборки ---

construct_items = {
    'PU':   cols[13:16],
    'PEOU': cols[16:19],
    'SI':   cols[19:22],
    'FC':   cols[22:25],
    'HM':   cols[25:28],
    'PV':   cols[28:31],
    'HAB':  cols[31:34],
    'BI':   cols[34:37],
}

construct_names = {
    'PU': 'Воспринимаемая полезность',
    'PEOU': 'Лёгкость использования',
    'SI': 'Социальное влияние',
    'FC': 'Поддерживающие условия',
    'HM': 'Гедоническая мотивация',
    'PV': 'Ценность цены',
    'HAB': 'Привычка',
    'BI': 'Поведенческое намерение'
}

constructs = ['PU', 'PEOU', 'SI', 'FC', 'HM', 'PV', 'HAB', 'BI']

for name, items in construct_items.items():
    df[name] = df[items].mean(axis=1)

da = df.dropna(subset=constructs).copy()

print(f'Полная выборка: {len(df)}')
print(f'Аналитическая подвыборка UTAUT2: {len(da)}')
print(f'Исключено по ветвлению анкеты: {len(df) - len(da)}')
```

```python
da.to_excel(SUBSAMPLE_PATH, index=False)
print(f'Сохранено: {SUBSAMPLE_PATH}')
print('Размер подвыборки:', da.shape)
```
